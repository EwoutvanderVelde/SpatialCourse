```python
import pandas as pd
import geopandas as gd
import fiona; help(fiona.open)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("data/crime.csv")
low_impact_df= df.loc[df["categoria_delito"]== "DELITO DE BAJO IMPACTO"]
high_impact_df= df.loc[df["categoria_delito"]!= "DELITO DE BAJO IMPACTO"]
```

```python
#Convert to geodataframe
gdf_low = gd.GeoDataFrame(
    low_impact_df, geometry = gd.points_from_xy(low_impact_df.longitud, low_impact_df.latitud))
gdf_low.crs = "EPSG:4326"
gdf_low.to_crs(epsg = 4326)

gdf_high = gd.GeoDataFrame(
    high_impact_df, geometry=gd.points_from_xy(high_impact_df.longitud, high_impact_df.latitud))
gdf_high.crs = "EPSG:4326"
gdf_high.to_crs(epsg = 4326)
```

```python
colonias_gdf = gd.read_file("data/colonias.geojson", layer='colonias')
colonias_gdf["area"] =  colonias_gdf["geometry"].area
colonias_gdf["id"]  = colonias_gdf.index
colonias_gdf.to_csv("data/neighborhoods_polygons.csv")
colonias_gdf
```

```python
low_df = gdf_low.sjoin(colonias_gdf, how="right")
high_df = gdf_high.sjoin(colonias_gdf, how="right")
```

```python
#Low impact crime
crimes_low = low_df.groupby("id").size().to_frame('crimes').reset_index()
crimes_area_low = pd.merge(crimes_low,colonias_gdf, how = "inner", on = "id")
crimes_area_low["low_impact/area"] = (crimes_area_low["crimes"]/ crimes_area_low["area"])/100000
crimes_area_low.sort_values(by= "low_impact/area", ascending = False)
```

```python
#High impact crime
crimes_high = high_df.groupby("id").size().to_frame('crimes').reset_index()
crimes_area_high = pd.merge(crimes_high,colonias_gdf, how = "inner", on = "id")
crimes_area_high["high_impact/area"] = (crimes_area_high["crimes"]/ crimes_area_high["area"])/100000
crimes_area_high.sort_values(by= "high_impact/area", ascending = False)
```

```python
#merging
df_crime_final = crimes_area_high.copy()
df_crime_final["low_impact/area"]= crimes_area_low["low_impact/area"]

df_crime_final
```

```python
#Exporting to gpkg
df_crime_final = gd.GeoDataFrame(df_crime_final)
df_crime_final.to_file("data/crimes_zone.gpkg", driver="GPKG")
```

```python
#Spatial join with houses data
#Spatial join of houses data points and crime levels
df = pd.read_csv("data/EN_merged_mexico_houses_data.csv")
# Convert to geodataframe
gdf_houses = gd.GeoDataFrame(df, geometry=gd.points_from_xy(df.longitude, df.latitude))
gdf_houses.crs = "EPSG:4326"

#Spatial join
gdf = gdf_houses.sjoin(df_crime_final[["high_impact/area","low_impact/area", "geometry"]], how="left")
gdf.drop("index_right",axis = 1,  inplace = True)
```

```python
# Drop properties that do not have criminality rate because they could not be matched with a location in CDMX
gdf = gdf[~((gdf["low_impact/area"].isna()) | (gdf["high_impact/area"].isna()))]
```

```python
gdf.to_csv("data/houses_crime.csv")
gdf
```
